```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Exemple de données
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [1, 4, 2, 5, 3]

# Création de la figure et de l'axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Création du scatter plot 3D avec une couleur unique
ax.scatter(x, y, z, color='blue', edgecolor='black', alpha=1.0)  # Utilisation d'une couleur unique pour tous les points

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot with Uniform Color')

plt.show()

```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.view_init(azim=-90, elev=90)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

#off
ax.axis('off')

#on
ax.axis('on')
```
